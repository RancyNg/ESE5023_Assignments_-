```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cite:https://blog.csdn.net/Asher117/article/details/84346073/

wind= pd.read_csv('2281305.csv')
wind['WND1']= wind['WND'].map(lambda x:x.split(',')[3])
wind['month']= wind['DATE'].map(lambda x:x.split('-')[1])
wind['year']= wind['DATE'].map(lambda x:x.split('-')[0])
wind['WND2']=wind['WND1'].astype('int')
lst=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
for a in lst:
    
    table=wind.loc[(wind['year'] == a )].groupby(['month'])['WND2'].mean().reset_index()
    plt.plot(table['month'],table['WND2'])
    plt.legend(labels=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020'],fontsize=7)

plt.savefig(fname='Wind')


```
